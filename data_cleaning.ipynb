```python
import pandas as pd
```

```python
df = pd.read_csv("student_performance_data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x=df['StudyTimeWeekly'])
plt.title("Boxplot of Study Time Weekly")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['GPA'], bins=20, kde=True)

plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Number of Students")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)

plt.title("Study Time vs GPA")
plt.xlabel("Study Time Weekly (Hours)")
plt.ylabel("GPA")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='GradeClass', data=df)

plt.title("Distribution of Grade Classes")
plt.xlabel("Grade Class")
plt.ylabel("Number of Students")

plt.show()
```
